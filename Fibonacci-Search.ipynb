```python
#importing required libraries
import math
```

```python
def fibonacci_search(function,lower_limit_of_interval = 0,upper_limit_of_interval = 1,critical_interval_length = 0.0005):
  
  L0 = upper_limit_of_interval - lower_limit_of_interval
  
  def generate_fibo_seq(critical_interval_length):
    list_fibo = list()
    a=1
    b=1
    list_fibo.append(a)
    list_fibo.append(b)
    while (a + b) <= (L0/critical_interval_length):
      c = a + b
      list_fibo.append(c)
      a = b
      b = c
    return list_fibo

  #fibonacci search method
  fibo_seq = generate_fibo_seq(critical_interval_length)
  no_of_iterations = len(fibo_seq)

  for i in range(no_of_iterations-1):
    print("********Iteration {}*********".format(i+1))
    print("Interval: [{},{}]".format(lower_limit_of_interval,upper_limit_of_interval))

    x1 = lower_limit_of_interval + fibo_seq[-1-(i+1)]*(L0/(fibo_seq[-1]+fibo_seq[-2]))
    print("X1 : {}".format(x1))
    value_at_x1 = function(x1)
    print("Value at X1 : {}".format(value_at_x1))

    x2 = upper_limit_of_interval - fibo_seq[-1-(i+1)]*(L0/(fibo_seq[-1]+fibo_seq[-2]))
    print("X2 : {}".format(x2))
    value_at_x2 = function(x2)
    print("Value at X2 : {}".format(value_at_x2))


    if value_at_x2 > value_at_x1:
      lower_limit_of_interval = lower_limit_of_interval
      upper_limit_of_interval = x2
    else:
      lower_limit_of_interval = x1
      upper_limit_of_interval = upper_limit_of_interval

    print("")

  optimized_value_occurs_at = (lower_limit_of_interval+upper_limit_of_interval)/2
  return optimized_value_occurs_at
```

```python
def function_to_be_minimized(value_at):
  return 6*(math.exp(-2*value_at)) + 2*value_at*value_at

x = fibonacci_search(function_to_be_minimized,0,1,0.0005)
print("minimum occurs at : {}".format(x))
print("value at minimum point : {}".format(function_to_be_minimized(x)))
```
