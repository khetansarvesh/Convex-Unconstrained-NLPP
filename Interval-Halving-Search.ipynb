```python
import math   
```

```python
def function_value (value_at):
  return 6*(math.exp(-2*value_at)) + 2*value_at*value_at
```

```python
lower_limit_of_interval = 0
upper_limit_of_interval = 1
delta = 0.0001
critical_interval = 0.0005
no_of_iterations = 100
```

```python
#dichromotomous search method
for i in range(no_of_iterations+1):
  print("********Iteration {}*********".format(i))
  print("Interval: [{},{}]".format(lower_limit_of_interval,upper_limit_of_interval))
  L0 = upper_limit_of_interval - lower_limit_of_interval

  x0 = lower_limit_of_interval + L0/2
  print("x0 : {}".format(x0))
  value_at_x0 = function_value(x0)

  x1 = lower_limit_of_interval + (x0 - lower_limit_of_interval)/2
  print("x1 : {}".format(x1))
  value_at_x1 = function_value(x1)

  x2 = x0 + (x0 - lower_limit_of_interval)/2
  print("x2 : {}".format(x2))
  value_at_x2 = function_value(x2)

  if value_at_x2>value_at_x0 and value_at_x0>value_at_x1:
    lower_limit_of_interval = lower_limit_of_interval
    upper_limit_of_interval = x0
  if value_at_x2<value_at_x0 and value_at_x0<value_at_x1:
    lower_limit_of_interval = x0
    upper_limit_of_interval = upper_limit_of_interval
  if value_at_x1>value_at_x0 and value_at_x2>value_at_x0 and value_at_x2>value_at_x1:
    lower_limit_of_interval = x1
    upper_limit_of_interval = x2
  if value_at_x1>value_at_x0 and value_at_x2>value_at_x0 and value_at_x1>value_at_x2:
    lower_limit_of_interval = x2
    upper_limit_of_interval = x1

  if (upper_limit_of_interval - lower_limit_of_interval) <= critical_interval:
    break

  print("")
```

```python
print("minimum occurs at : {}".format(lower_limit_of_interval))
print("value at minimum point : {}".format(function_value(lower_limit_of_interval)))
```
