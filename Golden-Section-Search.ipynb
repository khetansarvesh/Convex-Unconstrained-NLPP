```python
  #importing required libraries
  import math
```

```python
def golden_section_search (function,lower_limit_of_interval = 0,upper_limit_of_interval = 1,critical_interval_length = 0.0005):

  golden_ratio = 1.618
  L0 = upper_limit_of_interval - lower_limit_of_interval
  no_of_iterations = math.ceil(math.log2(L0/critical_interval_length)/math.log2(golden_ratio) +1 )

  #converging to the optimized solu tion
  for i in range(no_of_iterations-1):
    print("********Iteration {}*********".format(i+1))
    print("Interval: [{},{}]".format(lower_limit_of_interval,upper_limit_of_interval))

    x1 = lower_limit_of_interval + (1/(math.pow(golden_ratio,i+2)))*(L0)
    print("X1 : {}".format(x1))
    value_at_x1 = function(x1)
    print("Value at X1 : {}".format(value_at_x1))

    x2 = upper_limit_of_interval - (1/(math.pow(golden_ratio,i+2)))*(L0)
    print("X2 : {}".format(x2))
    value_at_x2 = function(x2)
    print("Value at X2 : {}".format(value_at_x2))

    if value_at_x2 > value_at_x1:
      lower_limit_of_interval = lower_limit_of_interval
      upper_limit_of_interval = x2
    else:
      lower_limit_of_interval = x1
      upper_limit_of_interval = upper_limit_of_interval

    print("")

  optimized_value_occurs_at = (lower_limit_of_interval+upper_limit_of_interval)/2
  return optimized_value_occurs_at
```

```python
#function to be minimized
def function_to_be_minimized(x):
  return 6*(math.exp(-2*x)) + 2*x*x

x = golden_section_search(function_to_be_minimized,0,1,0.0005)
print("minimum occurs at : {}".format(x))
print("value at minimum point : {}".format(function_to_be_minimized(x)))
```
