```python
def bisection_search_algorithm(function,function_first_derivative,lower_limit_of_interval=0,upper_limit_of_interval=1,critical_interval_length=0.0005):

  #calculating no of iterations required to converge
  no_of_iterations = math.ceil(math.log2((upper_limit_of_interval-lower_limit_of_interval)/critical_interval_length))
  
  #convering to the optimized solution
  for i in range(no_of_iterations):
    print("*********Iteration {}**********".format(i+1))
    print("Interval: [{},{}]".format(lower_limit_of_interval,upper_limit_of_interval))
    L0 = upper_limit_of_interval - lower_limit_of_interval

    x = lower_limit_of_interval + (L0/2)
    print("X : {}".format(x))
    slope_at_x = function_first_derivative(x)
    print("Slope at X : {}".format(slope_at_x))

    if slope_at_x > 0:
      lower_limit_of_interval = lower_limit_of_interval
      upper_limit_of_interval = x
    else:
      lower_limit_of_interval = x
      upper_limit_of_interval = upper_limit_of_interval
      
    print("")

  optimized_value_occurs_at = (lower_limit_of_interval+upper_limit_of_interval)/2
  return optimized_value_occurs_at
```

```python
#This algorithm is specifically for finding minima of 6e^(-2x) + 2x^2.........change this function if you want to find minima of some other function 
def function_to_be_minimized(x):
  return 6*math.exp(-2*x) + 2*x*x

#This algorithm is specifically for finding minima of 6e^(-2x) + 2x^2.........change this derivate if you want to find minima of some other function 
def first_derivative_of_function_to_be_minimized(x):
  return (-12)*math.exp(-2*x) + 4*x

x = bisection_search_algorithm(function_to_be_minimized,first_derivative_of_function_to_be_minimized,0,1,0.0005)
print("minimum occurs at : {}".format(x))
print("value at minimum point : {}".format(function_to_be_minimized(x)))
```
