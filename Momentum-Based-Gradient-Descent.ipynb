```python
import math
```

```python
def momentum_based_gradient_descent_algorithm(function , first_derivate , initial_random_start_at = 0.8 , threshold_value = 0.0005 , learning_rate = 0.1 , momentum = 0.9):

  current_point = initial_random_start_at
  new_point = current_point + 100000000
  i = 0
  update = 0

  #using difference between two values of x as stopping criteria
  while abs(current_point - new_point) > threshold_value:
    if i != 0:
      current_point = new_point
    
    print("*********Iteration {}**********".format(i+1))
    print("Current Point : {}".format(current_point))

    first_derivative_at_current_point = first_derivate(current_point)
    print("First Derivative At Current Point : {}".format(first_derivative_at_current_point))

    update = momentum*update + learning_rate*first_derivative_at_current_point
    new_point = current_point - update
    print("New Point : {}".format(new_point))

    i=i+1
    print("")
  
  optimum_occurs_at = new_point
  return optimum_occurs_at
```

```python
#This algorithm is specifically for finding minima of 6e^(-2x) + 2x^2.........change this function if you want to find minima of some other function 
def function(x):
  return 6*math.exp(-2*x) + 2*x*x

#This algorithm is specifically for finding minima of 6e^(-2x) + 2x^2.........change this derivate if you want to find minima of some other function 
def first_derivative(x):
  return (-12)*math.exp(-2*x) + 4*x

x = momentum_based_gradient_descent_algorithm(function,first_derivative,0.7,0.0005,0.01,0.9)
print("Minima Occurs At : {}".format(x))
print("Value At Minima : {}".format(function(x)))
```
